```python
import scipy
import torch
import albumentations as A
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader, Sampler
from skimage.io import imread

train_pipe = A.Sequential([
    A.LongestMaxSize(max_size=512),
    A.PadIfNeeded(min_height=512, min_width=512),
    A.ShiftScaleRotate(),
    A.RandomBrightnessContrast(),
])

eval_pipe = A.Sequential([
    A.LongestMaxSize(max_size=512),
    A.PadIfNeeded(min_height=512, min_width=512),
])

ages = [ idx for idx in range(3, 27, 3) ] + [ 24, 28, 32, 36 ] + [ idx for idx in range(42, 19*12+6, 6)]
tree = scipy.spatial.KDTree([ np.array((a, )) for a in ages ])

a2c = { idx: tree.query(np.array((idx, )))[1] for idx in range(1, 19*12+1) }

class WristDataset(Dataset):
    def __init__(self, names, desc, pipe, training=False):
        self.names = names
        self.desc = desc
        
        self.pipe = pipe

        self.training = training
        
    def __len__(self):
        return len(self.names)

    def __getitem__(self, idx):
        name = self.names[idx]
        img = imread(Path('boneage-training-dataset') / (str(name) + '.png'))

        img = self.pipe(image=img)['image']
        img = torch.tensor(img)[None]
        img = img.repeat(3, 1, 1).float()

        age = data[data['id'] == int(name)]['boneage'].iloc[0]
        class_ = a2c[age]

        return img, class_ if self.training else age

guys_names = data[data['male'] == True]['id'].tolist()
guys_ages = data[data['male'] == True]['boneage']

guys_classes = [ a2c[idx] for idx in guys_ages ]

girls_names = data[data['male'] == True]['id'].tolist()
girls_ages = data[data['male'] == True]['boneage']

girls_classes = [ a2c[idx] for idx in girls_ages ]

X_guys_train, X_guys_test, y_guys_train, y_guys_test = train_test_split(
guys_names, guys_ages, test_size=0.15, random_state=42)

X_girls_train, X_girls_test, y_girls_train, y_girls_test = train_test_split(
girls_names, girls_ages, test_size=0.15, random_state=42)

X_guys_train, X_guys_val, y_guys_train, y_guys_val = train_test_split(
X_guys_train, y_guys_train, test_size=0.15 * 0.85, random_state=42)

X_girls_train, X_girls_val, y_girls_train, y_girls_val = train_test_split(
X_girls_train, y_girls_train, test_size=0.15 * 0.85, random_state=42)


plt.hist(y_guys_train, alpha=0.5)
plt.hist(y_guys_val, alpha=0.5)
plt.hist(y_guys_test, alpha=0.5)
```

```python
import timm
```

```python
import timm

from tqdm import tqdm
from pathlib import Path

import torch.optim.swa_utils as tsu

def get_ema_avg_fn(decay=0.99):
    @torch.no_grad()
    def ema_update(ema_param, current_param, num_averaged):
        return decay * ema_param + (1 - decay) * current_param

    return ema_update

model = timm.create_model('tiny_vit_5m_224.dist_in22k', pretrained=True, num_classes=44)
model.cuda()

averaged_model = tsu.AveragedModel(model, avg_fn=get_ema_avg_fn())
averaged_model.cuda()

opt = torch.optim.AdamW(model.parameters(), lr=1e-3)


train_dataset = WristDataset(X_guys_train+X_girls_train, data, train_pipe, training=True)
val_dataset = WristDataset(X_guys_val+X_girls_val, data, eval_pipe)
test_dataset = WristDataset(X_guys_test+X_girls_test, data, eval_pipe)


from collections import Counter


class CustomSampler(Sampler):
    def __init__(self, classes):
        self.classes = [a2c[c] for c in classes ] 

        counts = dict(Counter(self.classes))
        cprobs = { a: 1 / 44 for a in range(44) }

        self.probs = []

        for idx in self.classes:
            prob = cprobs[idx] / counts[idx]
            self.probs.append(prob)

        self.probs = [ p / sum(self.probs)for p in self.probs]

        self.size = len(classes)

    def __len__(self):
        return self.size

    def __iter__(self):
        for _ in range(self.size):
            idx = np.random.choice(self.size, p=self.probs)
            yield idx
                    
sampler = CustomSampler(y_guys_train.tolist() + y_girls_train.tolist())

train_dataloader = DataLoader(train_dataset, 8, num_workers=4,  sampler=sampler)
val_dataloader = DataLoader(val_dataset, 8, num_workers=4, shuffle=False)
test_dataloader = DataLoader(test_dataset, 8, num_workers=4, shuffle=False)

device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

c2a = {c:a for c, a in enumerate(ages)}

for _ in range(6):
    model.train()
    
    for imgs_batch, labels_batch in tqdm(train_dataloader):
        imgs_batch = imgs_batch.cuda() 
        labels_batch = labels_batch.cuda()
        
        logit_batch = model(imgs_batch)
        loss = torch.nn.functional.cross_entropy(logit_batch, labels_batch)
        
        loss.backward()

        loss = loss.detach().cpu().numpy()        
        assert not np.isnan(loss)

        opt.step()
        model.zero_grad()

        averaged_model.update_parameters(model)

    tsu.update_bn(train_dataloader, averaged_model, device=device)
    
    true = list()
    pred = list()

    averaged_model.eval()
    
    for imgs_batch, labels_batch in tqdm(val_dataloader):
        with torch.no_grad():
            imgs_batch = imgs_batch.cuda() 
            labels_batch = labels_batch.cuda()

            logit_batch = averaged_model(imgs_batch)
            pred_batch = logit_batch.argmax(1)

            true.append(labels_batch.detach().cpu().numpy())
            pred.append(pred_batch.detach().cpu().numpy())

    true = np.concatenate(true)
    pred = [ c2a[e] for e in np.concatenate(pred) ]
    
    from sklearn.metrics import mean_absolute_error, mean_squared_error

    print(f'mae {mean_absolute_error(true, pred)}')
    print(f'mse {mean_squared_error(true, pred)**0.5}')
    

```

```python
true = list()
pred = list()

averaged_model.eval()

for imgs_batch, labels_batch in tqdm(test_dataloader):
    with torch.no_grad():
        imgs_batch = imgs_batch.cuda() 
        labels_batch = labels_batch.cuda()

        logit_batch = averaged_model(imgs_batch)
        pred_batch = logit_batch.argmax(1)

        true.append(labels_batch.detach().cpu().numpy())
        pred.append(pred_batch.detach().cpu().numpy())

true = np.concatenate(true)
pred = [ c2a[e] for e in np.concatenate(pred) ]

from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score

print(f'mae {mean_absolute_error(true, pred)}')
print(f'mse {mean_squared_error(true, pred)**0.5}')
```
